*Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.*

**Исходные данные имеют следующую структуру:**

**перечень всех документов**

` documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]` 

**перечень полок, на которых хранятся документы (если документ есть в documents, то он обязательно должен быть и в directories)**
` directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}`

Общие требования к программе:

код должен быть грамотно декомпозирован (каждая функция отвечает за свою конкретную задачу, дублирующийся функционал переиспользуется, а его код не повторяется);
в коде отсутствуют глобальные переменные (за исключением documents и directories);
пользовательский ввод обрабатывается в цикле while до тех пор, пока пользователь явно не завершит программу (вводом команды "q").

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
#______________________________________________________________функция вывода имени по номеру (пункт 1)
def get_name(documents): 
    num = input('Введите номер документа ')
    for i in documents:
        dictionary = i
        if dictionary['number'] == num:
            return dictionary['name']
    else:
        return ('Документ не найден в базе')
#______________________________________________________________ функция вывода номера полки (пункт 2)
def get_key(directories):
    num = input('Введите номер документа ')
    for k, v in directories.items():
        for i in v:
            if i == str(num):
                return k
#______________________________________________________________ функция вывода всей информации (пункт 3)
def all_info():
    for i in documents:
        dictionary = i
        number = dictionary['number']
        for k, v in directories.items():
            for i in v:
                if i == number:
                    dictionary.update({'branch':k})
        print ('№: ',dictionary['number'],', тип: ',dictionary['type'], ', владелец: ',dictionary['name'], ', полка номер: ', dictionary['branch'], sep='')
#______________________________________________________________ функция добавления полки (пункт 4)
def add_branch_number ():
    branch_add = 0
    number = input('Введите номер полки: ')
    if number in list(directories.keys()):
        print('Такая полка уже есть =(')
    else:
        directories.update({number:[]})
        print(f'Полка добавлена. Текущий перечень полок: {",".join(directories.keys())}')
#______________________________________________________________ функция удаления полки (пункт 5)
def del_branch():
    branch_del = 0
    number = input('Введите номер полки: ')
    if number in list(directories.keys()) and directories[number] == []:
        del directories[number]
        print(f'Полка удалена. Текущий перечень полок: {",".join(directories.keys())}')
    elif number in directories and directories[number] != []:
        print('Полка занята')
    else:
        print('Такой полки не существует')
        
def command_request():
    if command == ('l'):
        all_info()
    elif command == ('as'):
        add_branch_number ()
    elif command == ('ds'):
        del_branch()
    elif command == ('p'):
        print('Владелец документа:', get_name(documents))
    elif command == ('s'):
        print('Данные хранятся на полке:', get_key(directories))
    else:
        print('Неверная команда')
while True:
    command = input('Введите команду ')
    if command == ('q'):
        break
    else:
        command_request()
```
