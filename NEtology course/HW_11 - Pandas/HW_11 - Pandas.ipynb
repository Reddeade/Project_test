### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
import pandas as pd

power = pd.read_csv('power.csv')
Power_union = (power['country'] == 'Latvia') | (power['country'] == 'Lithuania') | (power['country'] =='Estonia')
Power_date = (power['category'] == 4) | (power['category'] == 12) | (power['category'] == 21)
sum_ = power[Power_union & Power_date & (power.quantity >= 0) & (power.year >= 2005) & (power.year <= 2010)]

print(f'Суммарное потребление стран Прибалтики: {sum_.quantity.sum()}')
```
